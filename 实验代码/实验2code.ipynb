```python
#############微博数据清洗
import pandas as pd
import os
import re
import warnings
warnings.filterwarnings("ignore")
```

```python
file = "../实验二数据/广东官微.xls"
df = pd.read_excel(file,index_col=0)
df
```

```python
columns = ['微博内容', '时间', '点赞数', '评论数', '转发数']
data1 = df[columns]
data1
```

```python
remove_ls = ['...','\u200b','O网页链接','\u3000']
for i in range(len(data1.index)):
    stri = str(data1.iloc[i,0])
    for j in remove_ls:
        stri = stri.replace(j, "")
    data1.iloc[i,0] = stri
data1
```

```python
####对时间进行处理，只保留年份，用transform（）函数
# data1.iloc[1,1][:4]
data1['时间'] = data1['时间'].transform(lambda x:x[:4])
data1
```

```python
####计算这个微博每一年的发文量
middf = data1.groupby('时间')['微博内容'].count()
middf.name = "年发文量"
middf
```

```python
data1 = pd.merge(data1, middf, on=['时间'], how='left')
data1
```

```python
mid_str = "知识产权、专利、商标、版权、著作权"
mid_ls = mid_str.split("、")
mid_ls
```

```python
###知识产权平均字数求解：
def knowlegde_count(dfi):
    count_all,count_knowlegde = 0, 0
    for i in range(len(dfi.index)):
        strii = dfi.iloc[i,0]
        count_all += len(strii)
        base = False
        for k in mid_ls:
            if strii.find(k) != -1 : base = True
        if base : count_knowlegde += len(strii)
    return count_knowlegde/count_all
```

```python
# s = knowlegde_count(data1[data1["时间"] == '2011'])
```

```python
time_ls = ['2011','2012','2013','2014','2015','2016','2017','2018','2019']
dicti = {timei:knowlegde_count(data1[data1["时间"] == timei]) for timei in time_ls}
print(dicti)
```

```python
a = pd.Series(dicti)
a.index.name = '时间'
a.name = '知识产权平均字数'
a
```

```python
data1 = pd.merge(data1, a, on=['时间'], how='left')
data1
```
